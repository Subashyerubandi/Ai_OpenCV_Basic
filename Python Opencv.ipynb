```python
cap=cv2.VideoCapture('99_preview_preview.mp4')
while(cap.isOpened()):
    ret, frame=cap.read()
    print(ret)
    if ret==True:
        cv2.imshow('frame',frame)
        if cv2.waitKey(200) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread('car.jpg')
bluring=cv2.blur(img,(10,10))
cv2.imshow('blurred image',bluring)

cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
import cv2

image=cv2.imread('birthday.jpeg')
start_point=(0,0)
end_point=(5000,5000)
color=(0,255,0)
thickness=9
img=cv2.line(image,start_point,end_point,color,thickness)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt
import cv2
image=cv2.imread('car.jpg')
center_coordinates=(460,270)
radius=250
color=(0,255,0)
thickness=2
img=cv2.circle(image,center_coordinates,radius,thickness)
img.shape
plt.imshow(img)
```
